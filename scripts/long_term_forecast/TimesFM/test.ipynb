```python
import os

# 【关键步骤】强制将 Hugging Face 端点指向国内镜像
# 这行代码必须放在 import timesfm 或 transformers 之前执行
os.environ['HF_ENDPOINT'] = 'https://hf-mirror.com'

# 建议同时忽略 SSL 验证（防止因为代理或证书问题导致的连接中断）
# os.environ['CURL_CA_BUNDLE'] = '' 

import numpy as np
import timesfm

# 初始化模型
model = timesfm.TimesFM_2p5_200M_torch.from_pretrained(
    "google/timesfm-2.5-200m-pytorch", 
    torch_compile=True
)

# 后续代码保持不变...
model.compile(
    timesfm.ForecastConfig(
        max_context=1024,
        max_horizon=256,
        normalize_inputs=True,
        use_continuous_quantile_head=True,
        force_flip_invariance=True,
        infer_is_positive=True,
        fix_quantile_crossing=True,
    )
)

point_forecast, quantile_forecast = model.forecast(
    horizon=12,
    inputs=[
        np.linspace(0, 1, 100),
        np.sin(np.linspace(0, 20, 67)),
    ], 
)

print("Point Forecast Shape:", point_forecast.shape)
print("Quantile Forecast Shape:", quantile_forecast.shape)
```
